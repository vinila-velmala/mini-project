```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("B5Customer_support_data.csv")
```

```python
# 1. Histogram of CSAT Score
plt.figure(figsize=(8,5))
sns.histplot(df['CSAT Score'], bins=10, kde=True, color="blue")
plt.title("CSAT Score Distribution")
plt.show()

# 2. Boxplot of CSAT Score
plt.figure(figsize=(6,4))
sns.boxplot(y=df['CSAT Score'], color="orange")
plt.title("CSAT Score Boxplot")
plt.show()

# 3. Violin Plot of CSAT Score
plt.figure(figsize=(6,4))
sns.violinplot(y=df['CSAT Score'], color="green")
plt.title("CSAT Score Violin Plot")
plt.show()

# 4. KDE Plot of CSAT Score
plt.figure(figsize=(8,5))
sns.kdeplot(df['CSAT Score'], shade=True, color="red")
plt.title("KDE of CSAT Score")
plt.show()

# 5. Bar Plot of CSAT Score Distribution
plt.figure(figsize=(8,5))
sns.countplot(x=df['CSAT Score'], palette="viridis")
plt.title("CSAT Score Count")
plt.show()
```

```python
# 6. Scatter Plot of Item_price vs CSAT Score
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Item_price'], y=df['CSAT Score'], color="purple")
plt.title("Item Price vs CSAT Score")
plt.show()

# 7. Scatter Plot of connected_handling_time vs CSAT Score
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['connected_handling_time'], y=df['CSAT Score'], color="brown")
plt.title("Connected Handling Time vs CSAT Score")
plt.show()
```

```python
# 8-15. Boxplots of CSAT Score with categorical variables
categorical_vars = ['category', 'Sub-category', 'Agent Shift', 'Tenure Bucket', 'Manager', 'Supervisor']
for var in categorical_vars:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[var], y=df['CSAT Score'])
    plt.xticks(rotation=45)
    plt.title(f"CSAT Score by {var}")
    plt.show()
```

```python
# 16. Bar Plot of CSAT Score by channel_name
plt.figure(figsize=(12,6))
sns.barplot(x=df['channel_name'], y=df['CSAT Score'])
plt.xticks(rotation=90)
plt.title("CSAT Score by Channel Name")
plt.show()
```

```python
# 17-19. Time Series Analysis
date_vars = ['order_date_time', 'Survey_response_Date', 'Issue_reported at']
for var in date_vars:
    df[var] = pd.to_datetime(df[var], errors='coerce')  # Convert to datetime
    df_time = df.groupby(var)['CSAT Score'].mean().dropna()
    plt.figure(figsize=(12,5))
    df_time.plot()
    plt.title(f"CSAT Score Over Time ({var})")
    plt.xlabel("Date")
    plt.ylabel("CSAT Score")
    plt.show()
```

```python
# 20. Heatmap of CSAT Score Correlation
plt.figure(figsize=(6,4))
sns.heatmap(df[['CSAT Score', 'Item_price', 'connected_handling_time']].corr(), annot=True, cmap="coolwarm")
plt.title("CSAT Score Correlation")
plt.show()

# 21. Pairplot of CSAT Score and Numerical Variables
sns.pairplot(df[['CSAT Score', 'Item_price', 'connected_handling_time']])
plt.show()

# 22. Swarmplot of CSAT Score by Category
plt.figure(figsize=(12,6))
sns.swarmplot(x=df['category'], y=df['CSAT Score'])
plt.xticks(rotation=45)
plt.title("CSAT Score by Category")
plt.show()
```

```python
# 27. Pie Chart of CSAT Score Distribution
plt.figure(figsize=(8,8))
df['CSAT Score'].value_counts().plot.pie(autopct='%1.1f%%', colormap="coolwarm")
plt.title("CSAT Score Distribution")
plt.show()

# 28. Boxen Plot of CSAT Score by Agent Name
plt.figure(figsize=(12,6))
sns.boxenplot(x=df['Agent_name'], y=df['CSAT Score'])
plt.xticks(rotation=90)
plt.title("CSAT Score by Agent Name")
plt.show()

# 29. ECDF Plot of CSAT Score
sns.ecdfplot(df['CSAT Score'])
plt.title("ECDF of CSAT Score")
plt.show()

# 30. Ridgeline Plot (KDE per category)
sns.displot(df, x="CSAT Score", hue="category", kind="kde", fill=True, height=6, aspect=2)
plt.title("CSAT Score by Category")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import FastICA
from sklearn.preprocessing import StandardScaler

# Selecting numerical columns for ICA (including CSAT Score)
num_features = ['CSAT Score', 'Item_price', 'connected_handling_time']
df_numeric = df[num_features].dropna()

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Apply FastICA with 3 components (adjust based on features)
ica = FastICA(n_components=3, random_state=42)
ica_transformed = ica.fit_transform(df_scaled)

# Convert to DataFrame
ica_df = pd.DataFrame(ica_transformed, columns=["ICA1", "ICA2", "ICA3"])
ica_df["CSAT Score"] = df_numeric["CSAT Score"].values  # Re-add CSAT Score





```

```python
# Plot 6: Pairplot of ICA Components with CSAT Score Hue
sns.pairplot(ica_df, vars=["ICA1", "ICA2", "ICA3"], hue="CSAT Score", palette="coolwarm")
plt.suptitle("Pairplot of ICA Components with CSAT Score", y=1.02)
plt.show()
```
